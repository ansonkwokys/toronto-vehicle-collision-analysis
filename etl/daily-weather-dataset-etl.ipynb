```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

import os
```

```python
daily_weather_df_array = []
for file in os.listdir("../toronto-weather-datasets-daily-2014-2025"):
    if file.endswith(".csv"):
        df = pd.read_csv(f"../toronto-weather-datasets-daily-2014-2025/{file}")
        daily_weather_df_array.append(df)
daily_weather_df = pd.concat(daily_weather_df_array)
daily_weather_df
        
```

```python
daily_weather_df.info()
```

Remove data after Jan 1 2025

```python
daily_weather_df = daily_weather_df[(daily_weather_df["Year"] != 2025) | ((daily_weather_df["Year"] == 2025) & (daily_weather_df["Month"] == 1) & (daily_weather_df["Day"] == 1))].copy()
```

```python
daily_weather_df.info()
```

```python
def list_all_column_values(df):
    for column_name, column in df.items():
        print(daily_weather_df.value_counts(column_name))
        print(" ")

list_all_column_values(daily_weather_df)
```

We would like to add an extra column for displaying the day in unix terms.

```python
from datetime import datetime
daily_weather_df["Datetime"] = 0
daily_weather_df["Datetime"] = df.apply(lambda row: int(datetime(row.Year, row.Month, row.Day, 0, 0, 0).timestamp() * 1000), axis=1)
daily_weather_df
```

There are missing data. We would like to investiage the values with N/A, starting from Max Temp.

```python
# fix null cell in max temperature
df_max_temp_null = daily_weather_df[daily_weather_df["Max Temp (°C)"].isna()]
df_max_temp_null
```
